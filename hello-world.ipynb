## Example 2 — SQL analytics task

```python
sql_task = TESTask.simple_sql(
    name="Mean Blood Pressure",
    query="""WITH user_query AS 
    (SELECT value_as_number FROM public.measurement WHERE measurement_concept_id = 3000905 AND value_as_number IS NOT NULL) 
    SELECT COUNT(*) AS n, SUM(value_as_number) 
    AS total FROM user_query;""",
)

sql_task.as_dict()
```

```python
result = sql_task.submit()
result
```
